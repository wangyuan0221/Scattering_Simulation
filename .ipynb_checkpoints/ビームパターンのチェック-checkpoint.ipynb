## header file の読み込み
ビームパターンの計算結果

```python
hfile = '/beampattern/mainlobe_zenith00mainlobe_azimuth000_subarray25_zenithX045VX0000186_zenithY045VY0000573/header_range000300.txt'
f = open(hfile)
header_data = f.read().split()

#
#header
#
dfile = header_data[1]
TX_NUM = int(header_data[3])
RX_NUM = int(header_data[5])
NX = int(header_data[7])
NY = int(header_data[9])
XMIN = float(header_data[11])
XMAX = float(header_data[13])
YMIN = float(header_data[15])
YMAX = float(header_data[17])
radar_range = float(header_data[19])
zenith = float(header_data[21])
azimuth = float(header_data[23])
#
#check
#
print("data file: {}".format(dfile))
print("TX subarrays: {}".format(TX_NUM))
print("RX elements: {}".format(RX_NUM))
print("grid X num: {}".format(NX))
print("grid X num: {}".format(NY))
print("X min: {} [m]".format(XMIN))
print("X max: {} [m]".format(XMAX))
print("Y min: {} [m]".format(YMIN))
print("Y max: {} [m]".format(YMAX))
print("range: {} [m]".format(radar_range))
print("main lobe zenith: {} [deg]".format(zenith))
print("main lobe azimuth: {} [deg]".format(azimuth))
```

## binary file の読み込み

```python
num1 = NX * NY
num2 = NX * NY * RX_NUM
head = ("head","<i")
tail = ("tail","<i")
data_type = np.dtype([head, ("E",[("real", "<"+str(num2)+"d"),("imag", "<"+str(num2)+"d")]), tail])
```

```python
# f = open(dfile, mode='rb')
# # data = np.fromfile(f, dtype=data_type)
# data = struct.unpack('f'*532890,f.read(4*532890))
# data = np.array(data,dtype=data_type)
# f.close()
# reflex = data["E"]["real"] + 1j*data["E"]["imag"]
# reflex = reflex.reshape(NY, NX, RX_NUM)
# print(reflex.shape)
```

```python
f = open(dfile, mode='rb')
data = np.fromfile(f, dtype=data_type)
f.close()
reflex = data["E"]["real"] + 1j*data["E"]["imag"]
reflex = reflex.reshape(NY, NX, RX_NUM)
print(reflex.shape)
```

## アンテナ座標の表示

```python
#
#TX
#
dfile = './tx_antpos.dat'
d_ant = np.loadtxt(dfile)
xx_tx = d_ant[:,0]
yy_tx = d_ant[:,1]
typ_tx = d_ant[:,2]
ax_tx = np.average(xx_tx)
ay_tx = np.average(yy_tx)
#
#RX
#
dfile = './rx_antpos.dat'
d_ant = np.loadtxt(dfile)
xx_rx = d_ant[:,0]
yy_rx = d_ant[:,1]
ax_rx = np.average(xx_rx)
ay_rx = np.average(yy_rx)

#
#figure
#
fs = 20
ms = 300

plt.figure(figsize=(7,7))
plt.scatter(xx_tx, yy_tx, label="TX")
plt.scatter(xx_rx, yy_rx, label="RX", color="r", marker='*', s=ms)
plt.title('Antenna position', fontsize=fs+5)
plt.xlabel('X(E-W) [m]', fontsize=fs)
plt.ylabel('Y(S-N) [m]', fontsize=fs)
plt.tick_params(labelsize=fs)
plt.axes().set_aspect('equal')
plt.legend(fontsize=fs)
plt.grid()
plt.show()
```

```python
#
#x, y
#


x = np.linspace(XMIN, XMAX, NX)
y = np.linspace(YMIN, YMAX, NY)
XX, YY = np.meshgrid(x,y)

#
#parameter
#
fs = 20
lim = [-1000,1000]
position = [5,6,2,4,8]

#
#figure
#
fig = plt.figure(figsize=(15,15))
plt.subplots_adjust(wspace=0.3)
for rx in range(reflex.shape[2]):
    power = 20*np.log10(np.abs(reflex[:,:,rx]))
    power = power - np.max(power)
    ax = fig.add_subplot(3,3,position[rx])
    im = ax.pcolormesh(XX, YY, power, cmap='tab20c', vmax=0, vmin=-40)
    cbar = fig.colorbar(im, ax=ax, aspect=20, pad=0.05, shrink=0.7)
    ax.set_title('RX idx:{}, $\\theta$={:.0f}, $\\phi$={:.0f}'.format(rx, zenith, azimuth), fontsize=fs)
    ax.set_aspect('equal')
    ax.set_xlabel('E-W[m]', fontsize=fs)
    ax.set_ylabel('S-N[m]', fontsize=fs)
    ax.set_xlim(lim)
    ax.set_ylim(lim)
    ax.tick_params(labelsize=fs-5)
    ax.grid()
plt.show()
```
